```python
import os
import pandas as pd
import numpy as np
import anndata as ad

import spatialdm as sdm
from spatialdm.datasets import dataset
import spatialdm.plottings as pl

import matplotlib.pyplot as plt
print("SpatailDM version: %s" %sdm.__version__)
```

```python
## first read in stuff and select only macrophage and tumors

all_count = pd.read_csv('../data/livcan_counts_y4_x69.csv')
all_meta = pd.read_csv('../../../arturito_local/final_R_part/liver/data/livcan_meta_y4_x69_withClusters_0327.csv')
all_raw = pd.read_csv('../data/livcan_raw_counts_y4_x69-1.csv')
all_meta.head()
```

```python
tm_meta = all_meta.loc[all_meta.snapres1.isin([0,4,5,6,7,8,9,10,12,14,16,17,23,24,25]),:]
tm_counts = all_count.loc[all_meta.snapres1.isin([0,4,5,6,7,8,9,10,12,14,16,17,23,24,25]),:]
tm_raw = all_raw.loc[all_meta.snapres1.isin([0,4,5,6,7,8,9,10,12,14,16,17,23,24,25]),:]

#tm_meta.shape
print([tm_meta.shape, tm_counts.shape, tm_raw.shape])
```

```python
import seaborn as sns 

# only macrophages and tumors, 25k ish cells
sns.scatterplot(data=tm_meta, x='x_slide_mm', y = 'y_slide_mm', s = 1)
```

```python
dm_adata = ad.AnnData((tm_raw))
dm_adata.raw = dm_adata
dm_adata.X = tm_counts
dm_adata.obs['snapcluster'] = tm_meta.snapres1.values # add cluster
dm_adata.obsm['spatial'] = np.column_stack((tm_meta.x_slide_mm, tm_meta.y_slide_mm)) # add global location
dm_adata
```

```python
# check l value, take couple minutes, peak mem is 60GB
sdm.weight_matrix(dm_adata, l=2, n_neighbors = 30, single_cell=True) # not using single cell here since not sure the behavier or radius
```

```python
# visualize range of interaction
import seaborn as sns
plt.figure(figsize=(6,6))

sns.scatterplot(x = dm_adata.obsm['spatial'][:,0], y = dm_adata.obsm['spatial'][:,1], 
            hue = dm_adata.obsp['weight'].A[20], s = 5)
```

```python
## range look good enough
sdm.extract_lr(dm_adata, 'human', min_cell=10)  
```

### permutation selection

```python
import time
start = time.time()
sdm.spatialdm_global(dm_adata, 200, specified_ind=None, method='permutation', nproc=1)     # global Moran selection
sdm.sig_pairs(dm_adata, method='permutation', fdr=True, threshold=0.1)     # select significant pairs
print("%.3f seconds" %(time.time()-start))
```

```python
start = time.time()
sdm.spatialdm_local(dm_adata, n_perm=200, method='permutation', specified_ind=None, nproc=1)     # local spot selection
sdm.sig_spots(dm_adata, method='permutation', fdr=False, threshold=0.1)     # significant local spots
print("%.3f seconds" %(time.time()-start))
```
